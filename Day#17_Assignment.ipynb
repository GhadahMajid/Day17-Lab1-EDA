```python
# import warnings
# warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
df['EMPLOYMENT_TYPE'] = df['EMPLOYMENT_TYPE'].fillna('UNKNOWN')
df['EMPLOYMENT_TYPE'].value_counts()
```

```python
y = df["CREDIT_HISTORY_LENGTH"].str.extract("(\d+)yrs", expand=False).astype(int)
m = df["CREDIT_HISTORY_LENGTH"].str.extract("(\d+)mon", expand=False).astype(int)
df["CREDIT_HISTORY_LENGTH_MONTHS"] = y * 12 + m
y = df["AVERAGE_ACCT_AGE"].str.extract("(\d+)yrs", expand=False).astype(int)
m = df["AVERAGE_ACCT_AGE"].str.extract("(\d+)mon", expand=False).astype(int)
df["AVERAGE_ACCT_AGE_MONTHS"] = y * 12 + m
df.drop(columns=['AVERAGE_ACCT_AGE','CREDIT_HISTORY_LENGTH'], inplace=True)
df.head()
```

```python
from sklearn import preprocessing
df_categorical = df.select_dtypes(include=['object'])
df_categorical.head()

le = preprocessing.LabelEncoder()
df_categorical = df_categorical.apply(le.fit_transform)
df_categorical.head()

df = df.drop(df_categorical.columns, axis=1)
df = pd.concat([df, df_categorical], axis=1)
df.head()
```

```python
plt.figure(figsize = (20,15))
sns.heatmap(df.iloc[:,:20].corr(), annot=True);
```

```python
plt.figure(figsize = (30,20))
sns.heatmap(df.corr(), annot=True);
```
